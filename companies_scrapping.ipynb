```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36",
    "Accept-Language": "en-US,en;q=0.9",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8",
    "Referer": "https://www.google.com/",
}
webpage=requests.get("https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav",headers=headers)
```

```python
soup=BeautifulSoup(webpage.text,'lxml')
print(soup.prettify())
```

```python
company=soup.find_all("div",class_="companyCardWrapper")
name=[]
rating=[]
number_of_reviews=[]
information=[]
for i in company:
    name.append(i.find("h2").text.strip())
    rating.append(i.find("div",class_="rating_text rating_text--md" ).text.strip())
    number_of_reviews.append(i.find("span",class_="companyCardWrapper__ActionCount").text.strip())
    information.append(i.find("span",class_="companyCardWrapper__interLinking").text.strip())
```

```python
data={"name":name,"rating":rating,"number_of_reviews":number_of_reviews,"information":information}
info=pd.DataFrame(data)
info
```

```python
final_list = [] 

for j in range(1, 9):  
    url = "https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav&page={}".format(j)
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36",
        "Accept-Language": "en-US,en;q=0.9",
        "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8",
        "Referer": "https://www.google.com/",
    }

    webpage = requests.get(url, headers=headers)
    soup = BeautifulSoup(webpage.text, "lxml")
    company = soup.find_all("div", class_="companyCardWrapper")

    name = []
    rating = []
    number_of_reviews = []
    information = []

    for i in company:
        name.append(i.find("h2").text.strip())
        rating.append(i.find("div", class_="rating_text rating_text--md").text.strip())
        number_of_reviews.append(i.find("span", class_="companyCardWrapper__ActionCount").text.strip())
        information.append(i.find("span", class_="companyCardWrapper__interLinking").text.strip())

    data = {
        "name": name,
        "rating": rating,
        "number_of_reviews": number_of_reviews,
        "information": information,
    }
    info = pd.DataFrame(data)
    final_list.append(info)

final = pd.concat(final_list, ignore_index=True)
final
```

```python
final.to_csv("companies.csv",index=False)
```
